```python
import sys
import tensorflow as tf
tf.disable_v2_behavior()
tf.enable_eager_execution()
tf.enable_resource_variables() 
import os

print(tf.__version__)

from tensor2tensor.utils import registry
from tensor2tensor import problems
from tensor2tensor import models
from tensor2tensor.utils.trainer_lib import create_run_config, create_experiment
from tensor2tensor.utils.trainer_lib import create_hparams
```

```python
DATA_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/data") # This folder contain the data
TMP_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/tmp")
TRAIN_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/train") # This folder contain the model
EXPORT_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/export") # This folder contain the exported model for production
TRANSLATIONS_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/translation") # This folder contain  all translated sequence
EVENT_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/event") # Test the BLEU score
USR_DIR = os.path.expanduser("/content/gdrive/MyDrive/Machine_Translation/t2t/user") # This folder contains our data that we want to add
 
tf.gfile.MakeDirs(DATA_DIR)
tf.gfile.MakeDirs(TMP_DIR)
tf.gfile.MakeDirs(TRAIN_DIR)
tf.gfile.MakeDirs(EXPORT_DIR)
tf.gfile.MakeDirs(TRANSLATIONS_DIR)
tf.gfile.MakeDirs(EVENT_DIR)
tf.gfile.MakeDirs(USR_DIR)
```

```python
PROBLEM = "translate_envi_iwslt32k" # We chose a problem translation English to French with 32.768 vocabulary
MODEL = "transformer" # Our model
HPARAMS = "transformer_big_single_gpu" # Hyperparameters for the model by default 
                            # If you have a one gpu, use transformer_big_single_gpu
```

```python
t2t_problem = problems.problem(PROBLEM)
t2t_problem.generate_data(DATA_DIR, TMP_DIR)
```

```python
train_steps = 2000 # Total number of train steps for all Epochs
eval_steps = 100 # Number of steps to perform for each evaluation
batch_size = 1024
save_checkpoints_steps = 1000
ALPHA = 0.1
schedule = "continuous_train_and_eval"

# Init Hparams object from T2T Problem
hparams = create_hparams(HPARAMS)

# Make Changes to Hparams
hparams.batch_size = batch_size
hparams.learning_rate = ALPHA
#hparams.max_length = 256
```

```python
!t2t-trainer \
  --data_dir=$DATA_DIR \
  --problem=$PROBLEM \
  --model=$MODEL \
  --hparams_set=$HPARAMS \
  --hparams="batch_size=$batch_size" \
  --schedule=$schedule\
  --output_dir=$TRAIN_DIR \
  --train_steps=$train_steps \
  --worker-gpu=1 \ 
```

```python
SOURCE_TEST_TRANSLATE_DIR = TMP_DIR+"/tst2013.en"
REFERENCE_TEST_TRANSLATE_DIR = TMP_DIR+"/tst2013.vi"
BEAM_SIZE=1
```

```python
!t2t-translate-all \
  --source=$SOURCE_TEST_TRANSLATE_DIR \
  --model_dir=$TRAIN_DIR \
  --translations_dir=$TRANSLATIONS_DIR \
  --data_dir=$DATA_DIR \
  --problem=$PROBLEM \
  --hparams_set=$HPARAMS \
  --output_dir=$TRAIN_DIR \
  --t2t_usr_dir=$USR_DIR \
  --beam_size=$BEAM_SIZE \
  --model=$MODEL
```

```python
#Tính giá trị BLEU của model dựa trên 2 file test
!t2t-bleu \
   --translations_dir=$TRANSLATIONS_DIR \
   --model_dir=$TRAIN_DIR \
   --data_dir=$DATA_DIR \
   --problem=$PROBLEM \
   --hparams_set=$HPARAMS \
   --source=$SOURCE_TEST_TRANSLATE_DIR \
   --reference=$REFERENCE_TEST_TRANSLATE_DIR \
   --event_dir=$EVENT_DIR
```

```python
#export Model
!t2t-exporter \
  --data_dir=$DATA_DIR \
  --output_dir=$TRAIN_DIR \
  --problem=$PROBLEM \
  --model=$MODEL \
  --hparams_set=$HPARAMS \
  --decode_hparams="beam_size=1,alpha=$ALPHA" \
  --export_dir=$EXPORT_DIR
```
